```python
# realizar o carregamento dos dados
data = pd.read_csv(FILE_PATH)
data.head()
```

```python
# modificar os nomes das colunas
data.columns = data.columns.str.lower()
data.head()
```

```python
# analisando a distribuição original - drug
data['drug'].value_counts(normalize=True)
```

```python
# transformar o problema num problema binário (one vs all)
data['drug'] = data['drug'].apply(lambda x: 1 if x == 'DrugY' else 0)

# analisando a nova distribuição - drug
data['drug'].value_counts(normalize=True)
```

## Modelagem

- SVM

```python
# separar x e y
x = data.drop(columns=['drug'])
y = data[['drug']]

# separar treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
# criar um transformador de colunas categóricas
cat_transf = ColumnTransformer([
    ('ohe', OneHotEncoder(drop='first'), ['sex', 'bp', 'cholesterol'])
], remainder='passthrough')

# configurar o pipeline do modelo
svm = Pipeline([
    ('transf', cat_transf),
    ('scaler', RobustScaler()),
    ('model', SVC(random_state=2))
])
```

```python
# configura o K-Fold
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=2)

# configurar os parâmetros de busca
params_svc = {
    'model__kernel': ['rbf', 'sigmoid', 'cosine', 'linear'],
    'model__C': np.random.uniform(0.01, 10, 100),
    'model__gamma': np.random.uniform(0.01, 10, 100),
}

# configurar o buscador
random_search_svc = RandomizedSearchCV(
    estimator=svm,
    param_distributions=params_svc,
    n_iter=100,
    scoring='f1',
    cv=cv,
    refit=True,
    error_score=0
)
```

```python
# ajustar o modelo
random_search_svc.fit(x_train, y_train)
```

```python
# melhores parâmetros
print('SVM:', random_search_svc.best_params_)
```

```python
# melhor desempenho médio
print('SVM:', random_search_svc.best_score_)
```

```python
# analisando o desempenho final - SVM
# realizando novas previsões
yhat_train_svm = random_search_svc.best_estimator_.predict(x_train)
yhat_test_svm = random_search_svc.best_estimator_.predict(x_test)

# análise do desempenho
print('Desempenho - Base de Treino - SVM')
print(classification_report(y_train, yhat_train_svm))

print('Desempenho - Base de Teste - SVM')
print(classification_report(y_test, yhat_test_svm))
```
