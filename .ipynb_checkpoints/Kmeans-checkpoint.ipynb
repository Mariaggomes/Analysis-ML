The analysis aims to segment chocolates based on characteristics such as origin location, bean type, cocoa percentage, and ratings, using the K-Means clustering algorithm. After cleaning and preparing the data, the model was applied to form 30 clusters, whose cardinality and magnitude were evaluated to validate the quality of the groupings. Techniques such as data scaling and the elbow method were used to determine the optimal number of clusters. The analysis also included geographic visualizations with heat maps, revealing distribution patterns of manufacturers and cocoa bean origins.

```python
#  Imports 

import re
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import ticker as mtick


import folium
from folium.plugins import HeatMap
```

Fill the null (missing) values in the 'bean_type' column with the value 'Blend', which indicates a mix of cocoa beans, ensuring that no value remains empty:

```python
data["bean_type"] = data["bean_type"].fillna("Blend").astype(str)

data["bean_type"].unique()
```

Corrects typos, standardizes location names, and replaces abbreviations with full names in the `maker_location` column, ensuring consistency and accuracy of the chocolate makers' locations, preventing duplication or confusion in the data:

```python
errors = [
    ['Amsterdam', 'Netherlands'],
    ['Eucador', 'Ecuador'],
    ['Niacragua', 'Nicaragua'],
    ['Scotland', 'United Kingdom'],
    ['U.K.', 'United Kingdom'],
    ['Wales', 'United Kingdom'],
    ['U.S.A.', 'United States'],

]

for error in errors:
    data['maker_location'] = data['maker_location'].str.replace(error[0],
                                                                error[1],
                                                                regex=True)
```

The standardization and correction of spelling errors and abbreviations in the origin locations of chocolates ensure a consistent and clean naming convention. This process involves replacing abbreviations, separator characters, and correcting typos, as well as removing periods, which facilitates analysis by eliminating inconsistencies and unwanted variations in location names.

```python

def cleanup_spelling_abbrev(text):
    replacements = [
        ['-', ', '], ['/ ', ', '], ['/', ', '], ['\(', ', '], [' and', ', '], [' &', ', '], ['\)', ''],
        ['Dom Rep|DR|Domin Rep|Dominican Rep,|Domincan Republic', 'Dominican Republic'],
        ['Mad,|Mad$', 'Madagascar, '],
        ['PNG', 'Papua New Guinea, '],
        ['Guat,|Guat$', 'Guatemala, '],
        ['Ven,|Ven$|Venez,|Venez$', 'Venezuela, '],
        ['Ecu,|Ecu$|Ecuad,|Ecuad$', 'Ecuador, '],
        ['Nic,|Nic$', 'Nicaragua, '],
        ['Cost Rica', 'Costa Rica'],
        ['Mex,|Mex$', 'Mexico, '],
        ['Jam,|Jam$', 'Jamaica, '],
        ['Haw,|Haw$', 'Hawaii, '],
        ['Gre,|Gre$', 'Grenada, '],
        ['Tri,|Tri$', 'Trinidad, '],
        ['C Am', 'Central America'],
        ['S America', 'South America'],
        [', $', ''], [',  ', ', '], [', ,', ', '], ['\xa0', ' '],[',\s+', ','],
        [' Bali', ',Bali']
    ]
    for i, j in replacements:
        text = re.sub(i, j, text)
    return text

data['specific_origin'] = data['specific_origin'].str.replace('.', '', regex=False).apply(cleanup_spelling_abbrev)
```

# Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

Scale the selected numerical variables (cocoa_percent, rating, and the geographic coordinates) using the standardization technique with StandardScaler. This transforms the data so that it has a mean of 0 and a standard deviation of 1, ensuring that all variables are on the same scale.

```python
# Scaling input
scaler = StandardScaler()
columns_to_normalize = ["cocoa_percent",
                        "rating",
                        "maker_location_latitude",
                        "maker_location_longitude",
                        "broad_origin_latitude",
                        "broad_origin_longitude"]
data_to_scale = data[columns_to_normalize].copy()
data_to_scale = pd.DataFrame(scaler.fit_transform(data_to_scale),
                             columns=columns_to_normalize,
                             index=data_to_scale.index)
data_scaled = data.copy()
for col in columns_to_normalize:
    data_scaled[col] = data_to_scale[col]

data_scaled
```

Apply the K-Means algorithm to perform clustering on the scaled dataset (data_scaled), creating 30 clusters (defined by the variable k). The algorithm is initialized 100 times (n_init=100) to improve the results and allows up to 10,000 iterations to reach convergence.

```python
# Clustering
k = 30
seed = 42

kmeans = KMeans(n_clusters=k, n_init=100, max_iter=10000, random_state=seed)
kmeans.fit(data_scaled)
```

# Model Quality

```python
from collections import Counter
import numpy as np
from scipy.spatial.distance import euclidean

# Cardinality

clusters = kmeans.labels_

value_counts = sorted(Counter(clusters).items())
clusters_cardinality, cardinality = zip(*value_counts)
```

```python
centroids = kmeans.cluster_centers_


magnitude = []

for cluster in range(len(centroids)):
    indexes = np.where(clusters == cluster)[0]
    if len(indexes) == 0:
        magnitude.append(0)
        continue
    data_ = data.iloc[indexes, :]
    distances = data_.apply(lambda v: euclidean(v, centroids[cluster]), axis=1)
    magnitude.append(distances.sum())

```

The cardinality of each cluster is obtained by counting the number of elements in each group. The magnitude is calculated by summing the Euclidean distances of the points in each cluster to their centroid, which helps evaluate the compactness of each cluster.

```python

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

ax = axes[0]


sns.set_style('ticks')
ax.bar(clusters_cardinality, cardinality)
ax.yaxis.grid(True)
sns.despine(offset=5)
ax.set_title('Cardinality')
ax.set_xlabel('Cluster')
ax.set_ylabel('Count')


ax = axes[1]
ax.bar(range(len(centroids)), magnitude)
ax.yaxis.grid(True)
sns.despine(offset=5)
ax.set_title('Magnitude')
ax.set_xlabel('Cluster')
ax.set_ylabel('Count')

ax = axes[2]
ax.plot(cardinality, magnitude, 'x')
ax.set_ylabel('Magnitude')
ax.set_xlabel('Cardinality')
ax.set_title('Magnitude vs Cardinality')
```

They show that the clusters generated by K-Means vary significantly in cardinality, with some clusters containing more than 80 elements and others having fewer than 20. The magnitude, which represents the sum of the distances of the points to the centroid, also varies among the clusters, being greater in clusters with more elements. The relationship between magnitude and cardinality is nearly linear, indicating that larger clusters tend to be more dispersed.

# Number of Cluster

Aims to identify the ideal number of clusters (k) by training the K-Means algorithm with different values of k.

```python
from tqdm.auto import tqdm
n_ks = list(range(2, 80, 2))

results_clusters = []
for k in tqdm(n_ks):
    model = KMeans(n_clusters=k, n_init=100, max_iter=10000, random_state=seed)
    model.fit(data_scaled)
    results_clusters.append(model.inertia_)
```

```python
fig, ax = plt.subplots()
sns.set_style('ticks')
ax.plot(n_ks, results_clusters, 'o')
ax.grid(True)
sns.despine(offset=True)
```

Based on the curve, a value of k around 30 appears to be the appropriate choice, as it balances the reduction of inertia with the simplicity of the model.
